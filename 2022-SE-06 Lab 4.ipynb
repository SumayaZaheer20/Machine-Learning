```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/Dataset .csv")
df.head(2)
```

```python
# Create Histogram using Matplotlib
plt.figure(figsize=(8, 5))
plt.hist(df["Aggregate rating"], bins=10, color="blue", edgecolor="black")
# Add Labels and Title
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
# Show the plot
plt.show()
```

It helps you understand how restaurant ratings are distributed.
If most ratings are high (e.g., 4-5 stars), it indicates good quality restaurants.
If ratings are mostly low (e.g., 1-2 stars), it suggests poorly rated restaurants.
A balanced spread means ratings are evenly distributed.

```python
# 2. Count of Restaurants in Different Price Ranges
plt.figure(figsize=(6, 4))
sns.countplot(x="Price range", data=df, palette="coolwarm")
plt.title("Count of Restaurants in Different Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

This bar chart shows the count of restaurants across different price ranges, indicating that lower-priced restaurants (Price Range 1) are the most common, while higher-priced restaurants (Price Range 4) are the least frequent.

```python
# 3. Number of Restaurants Offering Online Delivery
plt.figure(figsize=(6, 4))
sns.countplot(x="Has Online delivery", data=df, palette="Set2")
plt.title("Restaurants Offering Online Delivery")
plt.xlabel("Has Online Delivery")
plt.ylabel("Number of Restaurants")
plt.show()
```

This bar chart shows the number of restaurants offering online delivery, indicating that a significantly higher number of restaurants do not provide online delivery compared to those that do.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Price range"], y=df["Aggregate rating"], palette="coolwarm")
plt.title("Price Range vs. Aggregate Rating")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

Expensive restaurants generally receive better ratings, but some budget-friendly places also perform well. However, cheaper restaurants have more inconsistent ratings.

```python
# Create a scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df["Votes"], df["Aggregate rating"], color="purple", alpha=0.5)
# Add title and labels
plt.title("Votes vs. Aggregate Rating")
plt.xlabel("Number of Votes")
plt.ylabel("Aggregate Rating")
# Show the plot
plt.show()
```

Popular restaurants with more votes tend to have higher ratings. However, many restaurants show high ratings even with 0 votes, suggesting a possible default rating system or a lack of user engagement

```python
# Count of cuisines
cuisine_count = df['Cuisines'].value_counts().head(10)  # Top 10 cuisines

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=cuisine_count.index, y=cuisine_count.values, palette='Set2')
plt.title('Top 10 Cuisines Offered by Restaurants')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.show()
```

This bar chart displays the number of restaurants offering different cuisines, showing that North Indian cuisine is the most popular, while Street Food and Bakery & Desserts have the least representation.

```python
# Count of restaurants with and without table booking
table_booking_count = df['Has Table booking'].value_counts()

# Plotting
plt.figure(figsize=(6, 6))
plt.pie(table_booking_count, labels=table_booking_count.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Proportion of Restaurants with Table Booking Options')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

This pie chart illustrates the proportion of restaurants that offer table booking options, showing that only 12.1% of restaurants provide this service, while the majority (87.9%) do not.

```python
# Plotting the distribution of votes
plt.figure(figsize=(8, 4))
sns.histplot(df['Votes'], bins=30, kde=True)
plt.title('Distribution of Votes for Restaurants')
plt.xlabel('Number of Votes')
plt.ylabel('Frequency')
plt.show()
```

This histogram shows that most restaurants have very few votes, with a sharp peak near zero. The distribution is right-skewed, meaning a small number of restaurants receive a very high number of votes. This suggests that only a few places are highly popular, while most get little customer engagement

```python
# 9. Count of Restaurants that Allow Table Booking
plt.figure(figsize=(6, 4))
sns.countplot(x="Has Table booking", data=df, palette="pastel")
plt.title("Restaurants Allowing Table Booking")
plt.xlabel("Has Table Booking")
plt.ylabel("Number of Restaurants")
plt.show()
```

This bar chart shows the number of restaurants that allow table booking, revealing that a significantly higher number of restaurants do not offer this service compared to those that do.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x="Price range", y="Aggregate rating", data=df, palette="magma")
plt.title("Price Range vs Aggregate Rating")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.show()
```

Higher-priced restaurants generally have better ratings, while lower-priced ones have more variation.
